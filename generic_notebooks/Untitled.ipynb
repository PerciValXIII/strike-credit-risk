```python
def split_numerical_categorical(df: pd.DataFrame):
    """
    Splits dataframe columns into numerical and categorical features.

    Parameters
    ----------
    df : pd.DataFrame

    Returns
    -------
    numerical_features : list
        Column names with numeric dtype
    categorical_features : list
        Column names with non-numeric dtype
    """
    numerical_features = df.select_dtypes(include=["number"]).columns.tolist()
    categorical_features = df.select_dtypes(exclude=["number"]).columns.tolist()

    return numerical_features, categorical_features
```

```python
num_cols, cat_cols = split_numerical_categorical(demog)

print("Numerical feature np:", len(num_cols))
print("Categorical features no:", len(cat_cols))
```

```python
num_cols, cat_cols = split_numerical_categorical(deq)

print("Numerical feature np:", len(num_cols))
print("Categorical features no:", len(cat_cols))
```

```python
num_cols, cat_cols = split_numerical_categorical(vin)

print("Numerical feature np:", len(num_cols))
print("Categorical features no:", len(cat_cols))
```
